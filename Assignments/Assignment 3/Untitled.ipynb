```python
def normalise_text(raw_text):
    lower_text = raw_text.lower()
    print(lower_text)
    
    lower_text_no_contractions = (lower_text
                                 .replace('’',"'")
                                 .replace('‘',"'")
                                  .replace('’',"'")
                                  .replace("'m"," am")
                                 .replace("ain't",'am not')
                                 .replace("can't",'can not')
                                 .replace("cannot",'can not')
                                  .replace("'ll", " will")
                                  .replace("'ve", "  have")
                                  .replace("'d", "  would")
                                  .replace("'s", "  is")
                                  .replace("n't", " not")
                                  .replace("'re",'are')
                                  .replace("&",'and'))
    
    
    lower_text_no_contractions_no_punctuation = (lower_text_no_contractions
                          .replace(",", "")
                          .replace("'", "")
                          .replace('"', "")
                          .replace(";", "")
                          .replace(".","")
                          .replace('’','')
                          .replace('‘','')
                          .replace('?','')
                            .replace('“','')
                          .replace("%"," percent")
                          .replace("percentage","percent"))


    lower_text_no_contractions_no_punctuation_no_digits = re.sub(r"(\d+)", lambda x: num2words.num2words(int(x.group(0))), lower_text_no_contractions_no_punctuation)


    
    tokens = nltk.word_tokenize(lower_text_no_contractions_no_punctuation_no_digits)
    filtered_text = [non_stop_word for non_stop_word in tokens if not non_stop_word in set(nltk.corpus.stopwords.words('english'))]
    print(filtered_text)
    
    
    return filtered_text
```

```python
def word_count(l):
    
    doc="Document {}"
    doc_index=0
    per_doc=dict()
    
    for string in l:
        counts = dict()
        words = string.split()

        for word in words:
            if word in counts:
                counts[word] += 1
            else:
                counts[word] = 1

        doc_index+=1
        per_doc[doc.format(doc_index)]=counts

    return per_doc

print( word_count(new_sentences))

def word_count_all(l):
    
    counts=dict()
    for string in l:
        words = string.split()

        for word in words:
            if word in counts:
                counts[word] += 1
            else:
                counts[word] = 1

    return counts
```

```python
import nltk
from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()
```

```python
finder = BigramCollocationFinder.from_documents(new_tweets)
finder.apply_freq_filter(2)
(finder.nbest(bigram_measures.pmi, 10))
```

```python
import math

vals=dict()

for bigram in finder.ngram_fd:
    w1 = bigram[0]  # first word
    w2 = bigram[1]  # second word
    o = finder.ngram_fd[(w1, w2)]  # frequency of (w1, w2) co-occurrence
    f1 = finder.word_fd[w1]        # marginal frequency of w1
    f2 = finder.word_fd[w2]        # marginal frequency of w2
    n = finder.N                   # total number of samples
    pmi = finder.score_ngram(bigram_measures.pmi, w1, w2)

    po, p1, p2 = (o/n), (f1/n), (f2/n)
   # print(f'w1={w1}, w2={w2}, f1={f1}, f2={f2}, co={o}, n={n}')
   # print(f'by formula log2({po}/{p1}*{p2}), pmi{bigram} =', math.log2(po/(p1*p2)))
    print(f'PMI{bigram} = {pmi}\n')
    
    vals[bigram]=pmi
```

```python
import math
def entropy(labels):
    freqdist = nltk.FreqDist(labels)
    probs = [freqdist.freq(l) for l in freqdist]
    return -sum(p * math.log(p,2) for p in probs)
```

```python
print("The spam set entropy is: {}".format(entropy(spam_set)))
print("The random set entropy is: {}".format(entropy(random_set)))
print("The spam and random set entropy is: {}".format(entropy(spam_set + random_set)))
```
